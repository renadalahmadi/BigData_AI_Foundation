```python
# Use loc to slice your DataFrame for the 2 columns that returned the largest sum
'''2010
   and
   2013
   have the largest number of immergrants
'''
# this funcation retrun the 2 col with the largest sum values
def top_cols_sum(df,ncols):
  sum = df.sum(numeric_only= True)
  largest = sum.sort_values(ascending=False, inplace=False).head(2)
  dic_of_sum = largest.to_dict()
  key_list = list(dic_of_sum.keys())
  return key_list

# using the func and storing the values in vari a,b
a,b = top_cols_sum(myData,2)

# using the loc
myData.loc[: ,[a,b]]
```

```python
# Use iloc to slice your DataFrame for the 2 columns that returned the largest mean
'''2010
    and
   2013
   have the largest number of immergrants and they also have the largest mean
'''

# this funcation retrun the 2 col with the largest mean
def top_cols_mean(df,ncols):
  mean = df.mean()
  largest = mean.sort_values(ascending=False, inplace=False).head(2)
  dic_of_mean = largest.to_dict()
  key_list = list(dic_of_mean.keys())
  return key_list

# using the func and storing the values in vari a,b
a,b = top_cols_mean(myData,2)

# matching the col with values
count = 0
for col in myData.columns:
    if (col == a):
     col_1 = count
    if (col == b):
     col_2 = count
    count += 1

# using the iloc
print(myData.iloc[: ,[col_1 , col_2]])
```
